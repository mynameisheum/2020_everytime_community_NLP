## 익명,시간,null값, 삭제된 댓글입니다.
행을 지우는 전처리


```
# 숫자 지우기

def clean_text(texts):
    pattern = r'[^가-힣\s]'
    text = re.sub(pattern,'', texts)
    return text

df.texts = df.texts.apply(clean_text)
df.texts
```

```
# 익명 지우기




def clean_text2(texts):
    pattern = r'익명'
    text = re.sub(pattern,'', texts)
    return text

df.texts = df.texts.apply(clean_text2)
df.texts
```

```
# 삭제된 댓글입니다 지우기


# def clean_text3(texts):
#     pattern = r'[삭제된 댓글입니다]'
#     text = re.sub(pattern,'', texts)
#     return text

# df.texts = df.texts.apply(clean_text3)
# df.texts
df.texts = df.texts.replace('삭제된 댓글입니다', '')
df.head(50)
```

## 금지어 설정

```
stop_words = '걔 그게 글 날 달 데 둘 애 얘 얘기 저 적 하나 전 난 일 걸 뭐 줄 만 건 분 개 끝 잼 이거 번 중 듯 때 게 내 말 나 수 거 점 것'
stop_words = stop_words.split(' ')
stop_words
```

## 금지어를 제외한 명사들

```
# 금지어 제외한 명사들

nouns = []
for text_list in texts_list:
    for noun in tagger.nouns(text_list):
        if noun not in stop_words:
            nouns.append(noun)

nouns[:10]
```

```
nouns[:100]
```

## 단어 빈도수 측정

```
from collections import Counter

nouns_counter = Counter(nouns)
top_nouns = dict(nouns_counter.most_common(50))

top_nouns
```

## 임베딩

```
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

def tokenize(d):
    return word_tokenize(d)
```

```
# list당 토큰화로 2차원으로 reshape

from konlpy.tag import Mecab 

tagger = Mecab()

nouns = []
for text_list in texts_list:
    nouns2 = []
    for noun in tagger.nouns(text_list):
        if noun not in stop_words:

            nouns2.append(noun)
    nouns.append(nouns2)

nouns[:10]
```

```
import numpy as np

drop_nouns = [index for index, sentence in enumerate(nouns) if len(sentence) <= 1]  # 1sentence가 1보다 작을때
nouns_text = np.delete(nouns, drop_nouns, axis = 0)
print(len(nouns_text))
```

```
nouns_text
```

## Gensim을 이용한 Word2Vec

## 임베딩 skip-gram

```
from gensim.models import Word2Vec

model = Word2Vec(sentences = nouns_text, window = 4,
                 size = 100, min_count=5, workers = 4, sg = 1)    # window = 몇개를 예쌍할것인가
                                                                    # size= 크기
                                                                    #workers  =병렬처리
                                                                    #sg = 0 => cbow

                            
                            # 보통 skip-gram이 좋다
                            # 한 단어를 보고 앞뒤 3개를 맞춰야하기때문!
```

```
model.wv.similarity('남친','책')
```

```
model.most_similar(positive =['술'])
```

```
model.wv.most_similar(positive = ['남친'], negative = ['여친'])
```

## 임베딩 시각화 - skip-gram

```
from gensim.models import KeyedVectors

model.wv.save_word2vec_format('love-w2v')
```
